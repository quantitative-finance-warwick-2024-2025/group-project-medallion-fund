```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import seaborn as sns
import os
source_path = os.getcwd()
output_path=source_path+'/'+'datafiles/'

sns.set_theme(style="darkgrid")

# Read the CSV files
path = output_path

asset_path=output_path+"/Assets.csv"
df_assets = pd.read_csv(asset_path, index_col=0, parse_dates=True)
df_assets.sort_index(inplace=True)

strategy_output_path=output_path+"/Strategy_Output.csv"
df_strategy = pd.read_csv(strategy_output_path)

if len(df_strategy) == len(df_assets):
    df_strategy.index = df_assets.index
else:
    print("Row count mismatch between Strategy_Output.csv and Assets.csv.")

df_strategy.sort_index(inplace=True)

# Plot the wealth evolution for all agents
plt.figure(figsize=(14, 7))
plt.plot(df_strategy.index, df_strategy['riskhater_wealth'], label='RiskHater Wealth',linewidth=2)
plt.plot(df_strategy.index, df_strategy['naive_wealth'], label='Naive Investor Wealth',linewidth=2)
plt.plot(df_strategy.index, df_strategy['markowitzsavvy1_wealth'], label='Markowitz Savvy1 (0.001) Wealth',linewidth=2)
plt.plot(df_strategy.index, df_strategy['markowitzsavvy2_wealth'], label='Markowitz Savvy2 (0.005) Wealth',linewidth=2)
#plt.plot(df_strategy.index, df_strategy['markowitzsavvy3_wealth'], label='Markowitz Savvy3 Wealth',linewidth=2)
plt.xlabel("Year")
plt.ylabel("Wealth")
plt.title("Wealth Evolution Across Agents")
plt.legend(loc="best")

# Axis format
ax = plt.gca()
ax.xaxis.set_major_locator(mdates.YearLocator())
ax.xaxis.set_major_formatter(mdates.DateFormatter('%y'))
plt.xticks()
plt.ylim([0, 8])
plt.tight_layout()
plt.show()

#Plot returns comparison 
plt.figure(figsize=(14, 7))
plt.plot(df_strategy.index,  df_strategy['riskhater_creturns'], label='RiskHater Returns', linewidth=2)
plt.plot(df_strategy.index,  df_strategy['naive_creturns'], label='Naive Investor Returns', linewidth=2)
plt.plot(df_strategy.index,  df_strategy['markowitzsavvy1_creturns'], label='Markowitz Savvy1 (0.001) Returns', linewidth=2)
plt.plot(df_strategy.index,  df_strategy['markowitzsavvy2_creturns'], label='Markowitz Savvy2 (0.005) Returns', linewidth=2)
#plt.plot(df_strategy.index,  df_strategy['markowitzsavvy3_creturns'], label='Markowitz Savvy3 Returns', linewidth=2)
plt.xlabel("Year")
plt.ylabel("Returns")
plt.title("Return Comparison Across Agents")
plt.legend(loc="best")
plt.ylim([-1,7])
ax = plt.gca()
ax.xaxis.set_major_locator(mdates.YearLocator())
ax.xaxis.set_major_formatter(mdates.DateFormatter('%y'))
plt.xticks()
plt.tight_layout()
plt.show()
```
