```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import seaborn as sns

# Set Seaborn theme for dark background and grid styling
sns.set_theme(style="darkgrid")
# plt.style.use("dark_background")

# --- Read the CSV files ---
# Read the assets file to extract the index (years)
path = "/Users/akshayzine/Academics/Programming_Group_Project/Akshay_Branch/group-project-medallion-fund/"

df_assets = pd.read_csv("Assets.csv", index_col=0, parse_dates=True)
df_assets.sort_index(inplace=True)

# Read the strategy output CSV (wealth and returns for agents)
df_strategy = pd.read_csv("Strategy_Output.csv")

# Assign the date index from assets to strategy data if row counts match.
if len(df_strategy) == len(df_assets):
    df_strategy.index = df_assets.index
else:
    print("Row count mismatch between Strategy_Output.csv and Assets.csv.")

df_strategy.sort_index(inplace=True)

# --- Plot the wealth evolution for all agents ---
plt.figure(figsize=(14, 7))
plt.plot(df_strategy.index, df_strategy['riskhater_wealth'], label='RiskHater Wealth',linewidth=2)
plt.plot(df_strategy.index, df_strategy['naive_wealth'], label='Naive Investor Wealth',linewidth=2)
plt.plot(df_strategy.index, df_strategy['markowitzsavvy1_wealth'], label='Markowitz Savvy1 Wealth',linewidth=2)
plt.plot(df_strategy.index, df_strategy['markowitzsavvy2_wealth'], label='Markowitz Savvy2 Wealth',linewidth=2)
#plt.plot(df_strategy.index, df_strategy['markowitzsavvy3_wealth'], label='Markowitz Savvy3 Wealth',linewidth=2)
plt.axhline(0)
plt.xlabel("Year")
plt.ylabel("Wealth")
plt.title("Wealth Evolution Across Agents")
plt.legend(loc="best")

# --- Format x-axis ticks ---
ax = plt.gca()
# Place ticks at the start of each year
ax.xaxis.set_major_locator(mdates.YearLocator())
# Format ticks to display last two digits of the year (e.g., "15", "16", etc.)
ax.xaxis.set_major_formatter(mdates.DateFormatter('%y'))
plt.xticks()
plt.ylim([0, 8])

plt.tight_layout()
plt.show()

# ---- Plot 2: Returns Comparison ----
plt.figure(figsize=(14, 7))
plt.plot(df_strategy.index,  df_strategy['riskhater_creturns'], label='RiskHater Returns', linewidth=2)
plt.plot(df_strategy.index,  df_strategy['naive_creturns'], label='Naive Investor Returns', linewidth=2)
plt.plot(df_strategy.index,  df_strategy['markowitzsavvy1_creturns'], label='Markowitz Savvy1 Returns', linewidth=2)
plt.plot(df_strategy.index,  df_strategy['markowitzsavvy2_creturns'], label='Markowitz Savvy2 Returns', linewidth=2)
#plt.plot(df_strategy.index,  df_strategy['markowitzsavvy3_creturns'], label='Markowitz Savvy3 Returns', linewidth=2)
plt.axhline(0)
plt.xlabel("Year")
plt.ylabel("Returns")
plt.title("Return Comparison Across Agents")
plt.legend(loc="best")
plt.ylim([0,7])

# --- Format x-axis ticks ---
ax = plt.gca()
ax.xaxis.set_major_locator(mdates.YearLocator())
ax.xaxis.set_major_formatter(mdates.DateFormatter('%y'))
plt.xticks()
plt.tight_layout()
plt.show()
```
